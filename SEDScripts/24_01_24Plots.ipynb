```python
from scipy.interpolate import interp1d, make_interp_spline
from sklearn.preprocessing import normalize
```

```python
# Proportional Controller
x = np.array(df[(df['Kd'] == 0) & (df['Ki'] == 0)]['Kp'])
y = np.array(df[(df['Kd'] == 0) & (df['Ki'] == 0)]['Pow50%'])
ST = np.array(df[(df['Kd'] == 0) & (df['Ki'] == 0)]['SetTime5%'])
POS = np.array(df[(df['Kd'] == 0) & (df['Ki'] == 0)]['PeakOS'])
norm1 = y / np.linalg.norm(y)
normST = ST / np.linalg.norm(ST)
normPOS = POS / np.linalg.norm(POS)                                                  
                                                     
cubic_interpolation_model = interp1d(x, norm1, kind ='quadratic')
cubic_interpolation_model2 = interp1d(x, normST, kind ='quadratic')
cubic_interpolation_model3 = interp1d(x, normPOS, kind ='quadratic')
# Plotting the Graph
X_=np.linspace(x.min(), x.max(), 500)
Y_=cubic_interpolation_model(X_)
Y2_=cubic_interpolation_model2(X_)
Y3_=cubic_interpolation_model3(X_)
plt.figure(figsize=(10,6))
plt.plot(X_, Y_, label='Power')
plt.plot(X_, Y2_, label='SettlingTime')
plt.plot(X_, Y3_, label='Peak Overshoot')
plt.title("Proportional Controller Parameters")
plt.xlabel("Kp")
plt.ylabel("Parameters")
plt.legend()
plt.show()
```
