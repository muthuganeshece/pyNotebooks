```python
df = pd.DataFrame(columns=['Kp', 'Kd', 'Ki', 'Pow20%', 'Pow50%','Pow60%','Pow90%', 'SetTime10%', 'SetTime5%', 'SetTime2%', 'PeakOS'])
df['Kp'] = TestKp
df['Kd'] = TestKd
df['Ki'] = TestKi
df.head()
```

```python
#fig, axs = plt.subplots(13, figsize=(15, 60))
#fig.suptitle('Vertically stacked subplots')
SettlingTime = []
SettlingTime_list = []
for i in range(len(TestKp)):
    TempData = raw_data[((raw_data['Kp'] == TestKp[i]) & (raw_data['Ki'] == TestKi[i]) & (raw_data['Kd'] == TestKd[i]))]
    SettlingTime_list = []
    PeakOverShoot = 0
    for j in range(1, len(PowP)):
        TempDataPow = TempData[TempData['SetPoint'] == PowP[j]][['Time','SetPoint','Power']]
        PeakOverShoot = PeakOverShoot + max(TempDataPow['Power'] - TempDataPow['SetPoint']) ** 2
        
        df[df.columns[j+2]].iloc[i] = round((((TempDataPow['Power'] ** 2) * 3).mean())/ (PowP[j] ** 2), 2)#20% power

        SettlingTime = []
        for k in range(len(SetT)):
            TempVar = TempDataPow[((TempDataPow['Power'] < ((1 + SetT[k]) * PowP[j])) & (TempDataPow['Power'] > ((1 - SetT[k]) * PowP[j])))][['Time','Power']]
            TempVar['Time'] = pd.to_datetime(TempVar['Time'],format='%H:%M:%S')
            TempVar['TimeDiff'] = TempVar['Time'].diff(10) // timedelta(seconds=1)
            
            if not TempVar['TimeDiff'].dropna().empty:
                SettlingTime.append(((TempVar[TempVar['TimeDiff'] >= 30]['Time'].iloc[0] - pd.to_datetime(TempDataPow['Time'].iloc[0],format='%H:%M:%S')) // timedelta(seconds=1)) / (PowP[j] - PowP[j-1]))
            else:
                SettlingTime.append(float("NAN"))
        if j != 0 and len(SettlingTime) != 0:    
            SettlingTime_list.append(SettlingTime)
        #print(PeakOverShoot)
    if len(SettlingTime_list) != 0:
        x = np.array(SettlingTime_list)
        df['SetTime10%'].iloc[i] = round(np.nanmean(x[:,0]), 2)
        df['SetTime5%'].iloc[i] = round(np.nanmean(x[:,1]), 2)
        df['SetTime2%'].iloc[i] = round(np.nanmean(x[:,2]), 2)
    #print(PeakOverShoot)
    PeakOverShoot = PeakOverShoot / (len(PowP) - 1)
    df['PeakOS'].iloc[i] = round(PeakOverShoot,2)
    
    
    
```
