# Data Cleaning

```python
df = raw_data.copy()
```

```python
df.loc[pd.isnull(raw_data['product_category']), 'product_category'] = 'Undefined'
```

```python
df.loc[pd.isnull(raw_data['product_name']), 'product_name'] = df.loc[pd.isnull(raw_data['product_name']), 'product_category']
```

```python
df.loc[pd.isnull(raw_data['product_price']), 'product_price'] = 5.8
```

```python
df = df.drop('discount_company', axis = 1)
```

```python
df.info()
```

```python
TempVar = df[['product_name', 'product_price']]
np.mean(TempVar.groupby(by = 'product_name').mean())
```

```python
df['sale_date'] = pd.to_datetime(df['sale_date'])
df.info()
```

```python
df['sale_time'] = pd.to_datetime(df['sale_date']).dt.time
df['sale_date'] = pd.to_datetime(df['sale_date']).dt.date
```
