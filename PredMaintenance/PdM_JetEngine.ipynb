```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_data = pd.read_csv('E:\\NextStep\\MLProjects\\YTS\\PredictiveMaintenance\\Datasets\\NASAJetEngine\\CMaps\\train_FD001.csv')
```

```python
raw_data
```

```python
cols = list(raw_data.columns)
for i in range(2,len(cols)):
    plt.figure(figsize=(12,6))
    plt.scatter(raw_data['Cycle'], raw_data.iloc[:,i])
    plt.title(cols[i])
```

```python
df = raw_data.drop(['Sensor3', 'Sensor4', 'Sensor8', 'Sensor13', 'Sensor19', 'Sensor21', 'Sensor22'], axis=1)
```

```python
df
```

```python
df_grp = df.groupby('Engine').mean()
df_grp.reset_index(inplace=True)
cols = list(df_grp.columns)
for i in range(2,len(cols)):
    plt.figure(figsize=(12,6))
    plt.scatter(df_grp['Engine'], df_grp.iloc[:,i])
    plt.title(cols[i])
```

```python
lst = list(df[['Engine', 'Cycle']].groupby('Engine').idxmax()['Cycle'])
temp_df = df.loc[lst]
cols = list(df.columns)
for i in range(2,len(cols)):
    plt.figure(figsize=(12,6))
    plt.scatter(df['Cycle'], df.iloc[:,i])
    plt.scatter(temp_df['Cycle'], temp_df.iloc[:,i], color='r')
    plt.title(cols[i])
```

```python
data = round(df.corr(), 2)
plt.figure(figsize=(12,6))
sns.heatmap(data[1:], annot = True)
```
